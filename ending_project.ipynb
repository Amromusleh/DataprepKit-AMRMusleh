```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

```python
def getting_file_input():
    file_name = input('put the file name here ')
    try:

        if file_name.find('.csv') != -1:
            data = pd.read_csv(file_name)

        else:
            print("make sure if the file name is corrct")
            getting_file_input()

        return data
    except:
        print("something went wrong")
        sys.exit()

data = getting_file_input()
```

Please provide the file name. If the file exists, it will be saved in the variable named 'data'





```python
def clean_data():
    global data

    cleand_data = data

    try:
        cleand_data.fillna(method='ffill', inplace=True)
        numeric_cols = cleand_data.select_dtypes(include=['number']).columns
        cleand_data[numeric_cols] = cleand_data[numeric_cols].apply(pd.to_numeric, errors='coerce')
        return cleand_data

    except :
        print(f"An error occurred while cleaning the data: ")
        return None

cleand_data = clean_data()
```

this function will clean the data from the file
by Converting numeric columns to numeric format and Filling missing values

```python
sorted_df = cleand_data.sort_values(by='bedrooms', ascending=True)
bedroom_counts = sorted_df['bedrooms'].value_counts().to_dict()
sorted_df['bedroom_count'] = sorted_df['bedrooms'].map(bedroom_counts)
point_size = sorted_df['bedroom_count'] * 2
sns.scatterplot(x='review_scores_rating', y='bedrooms', data=sorted_df, hue='room_type', size=point_size)
plt.xlabel('Review Scores Rating')
plt.ylabel('Bedrooms')
plt.title('Scatter Plot of Review Scores Rating and Bedrooms by Room Type')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

Based on the scatter plots, it's evident that more people tend to choose the entire home/apartment option. Additionally, the plots indicate that the prices are affordable even for the full home or apartment, which makes it an attractive option for guests.

```python

sorted_df = cleand_data.sort_values(by='availability_90', ascending=True)

sorted_df['price'] = sorted_df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

availability_freq = sorted_df['availability_90'].value_counts().to_dict()

sorted_df['availability_freq'] = sorted_df['availability_90'].map(availability_freq)

point_size = sorted_df['availability_freq'] * 2

sns.scatterplot(x='price', y='review_scores_rating', data=sorted_df, hue='availability_90', size=point_size)

plt.xlabel('Price')
plt.ylabel('Review Scores Rating')
plt.title('Scatter Plot of Price and Review Scores Rating')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

```

In the plot, we observe a trend where lower prices tend to correspond with higher review scores. This suggests that when a company offers good services at a reasonable price, it tends to attract more positive reviews from guests. It underscores the importance of providing excellent value for money, as it can significantly influence guest satisfaction and the overall reputation of the place.
